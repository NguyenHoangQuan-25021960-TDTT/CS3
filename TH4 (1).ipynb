```python
#W4A10
def do_dai(k):
    length = 1
    while k > 1:
        if k % 2 == 0:
            k = k // 2
        else:
            k = 3 * k + 1
        length += 1
    return length

n = int(input())

Max = 0
Num = 1

for i in range(1, n + 1):
    a = do_dai(i)
    
    if a > Max:
        Max = a
        Num = i
print(Num, Max)

        
        
    
    
    
        
    
```
