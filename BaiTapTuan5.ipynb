```python
#W5A1
def Max_Of_Two(a, b):
   if not a.lstrip('-').isdigit():
       print(" Nhap lai")
       return False
   if not b.lstrip('-').isdigit():
        print("Nhap lai")
        return False
       
   a = int(a)
   b = int(b)
   if a > b:
       return a
   else:
       return b
a, b = input().split()
print(Max_Of_Two(a, b))
```

```python
#W5A4
import math
def is_perfect_num(num):
    if num < 2:
        return False
    tong = 1
    for i in range(2, math.isqrt(num)+1):
        if num % i ==0:
            tong += i
            if i != num//i:
                tong += num//i
    return tong == num
def cac_so_hoan_hao(n):
    print(f" Cac so hoan hao từ 1 đến {n} là:")
    Flag = False
    for k in range(1,n+1):
        if is_perfect_num(k):
           print(k, end = " ")
           Flag = True
           
    if not Flag:
        print(" Khong có so hoan hao")
    print()
n = int(input())
cac_so_hoan_hao(n)

     
        
 
```

```python
#WA5
def list_integers(mylist,k):
    for i in range(len(mylist)):
        if mylist[i]==k:
           return i + 1
    return -1
mylist = list(map(int, input().split()))
k = int(input())
print(list_integers(mylist,k))
    
```

```python
#W5A6
def tinh_giai_thua(n):
    if not  n.lstrip("-").isdigit():
        print(" Nhap lai so")
    if n == 0 or n == 1:
        print("Giai thừa là: 1")
        
    n = int(n)
    giai_thua = 1
    for i in range(2, n+1):
        giai_thua = giai_thua * i
    print(f" Giai thua của {n} là:")
    return giai_thua
n = input()
print(tinh_giai_thua(n))
        
        
        
```

```python
#W5A7
def tinh_toan_tu(num_1,num_2, operat):
    if operat == "-":
        return round(num_1 - num_2, 2)
    elif operat == "+":
        return round(num_1-num_2, 2)
    elif operat == "/":
        return round(num_1/num_2, 2)
    elif operat == "*":
        if num_2 == 0:
            return "Lỗi"
        else:
             return round(num_1*num_2, 2)
    else:
        return " Toán tử không hợp lệ"
num_1,num_2 = map(float, input().split())
operat = input()
print(tinh_toan_tu(num_1,num_2,operat))
  
```
